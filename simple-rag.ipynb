```python
import os

folder_path = '/kaggle/input/physics'
md_files = [f for f in os.listdir(folder_path) if f.endswith('.md')]

all_content = {}
for file_name in md_files:
    with open(os.path.join(folder_path, file_name), 'r', encoding='utf-8') as f:
        all_content[file_name] = f.read()
```

```python
all_content
```

```python
all_topics = []
for chapter_name, content in all_content.items():
    topics = content.split('*****')
    for topic in topics:
        topic = topic.strip()
        if topic:
            # Optionally, you can remove the chapter name from the start of each topic
            # This is helpful for cleaner embeddings
            # topic_text = topic.replace(chapter_name.replace('.md', ''), '').strip()
            all_topics.append(topic)
```

```python
all_topics
```

```python
print(f"Total number of chunks created: {len(all_topics)}")
print("--- Printing the content of each chunk ---")

chunks = []

for i, chunk in enumerate(all_topics):
    # print(f"Chunk #{i+1}")
   
    print(chunk)
    chunks.append(chunk)
    # print("---------------------------------\n")
```

```python
import numpy as np
import os
from sentence_transformers import SentenceTransformer
import faiss
```

```python
embeddings = np.array(result['embedding'], dtype='float32')
print("Embeddings generated.")
```

```python
d = embeddings.shape[1] 
print(f"Creating FAISS index with dimension: {d}")
```

```python
# 'IndexFlatL2' is a simple index that uses L2 distance for similarity search.
index = faiss.IndexFlatL2(d) 
```

```python
# 3. Add the embeddings to the FAISS index
index.add(embeddings)
print("Embeddings added to FAISS index.")
```

```python
# 4. Create a simple dictionary to store the chunks,
# mapping the FAISS index ID to the original chunk text.
doc_store = {
    i: chunks[i] for i in range(len(chunks))
}
print("Chunks stored in the document store.")
```

```python
# To save the FAISS index
faiss.write_index(index, "my_faiss_index.faiss")

# To save the doc_store, you can use a library like pickle or json
import pickle

with open("my_doc_store.pkl", "wb") as f:
    pickle.dump(doc_store, f)
```

```python
# To load the FAISS index
loaded_index = faiss.read_index("my_faiss_index.faiss")

# To load the doc_store
with open("my_doc_store.pkl", "rb") as f:
    loaded_doc_store = pickle.load(f)

# Now you can use loaded_index and loaded_doc_store for search
```

```python
from rank_bm25 import BM25Okapi

# Tokenize your documents
tokenized_corpus = [doc.split(" ") for doc in chunks]

# Create the BM25 index
bm25 = BM25Okapi(tokenized_corpus)

print("BM25 index created successfully.")
```

```python
# Assuming 'query_text', 'loaded_index', 'loaded_doc_store', and 'bm25' are ready.
query_text = "পদার্থবিজ্ঞানের শাখা গুলো কি কি ?"


# 1. Semantic Search (FAISS)
k_search = 10 # Get more results to improve RRF's effectiveness
query_embedding = np.array(genai.embed_content(
    model="models/gemini-embedding-001",
    content=query_text
)['embedding'], dtype='float32').reshape(1, -1)
distances, semantic_indices = loaded_index.search(query_embedding, k_search)
semantic_indices = semantic_indices[0] # Flatten the result

# Create a dictionary to map doc index to its semantic rank
semantic_ranks = {doc_id: rank + 1 for rank, doc_id in enumerate(semantic_indices)}

# 2. Keyword Search (BM25)
tokenized_query = query_text.split(" ")
keyword_scores = bm25.get_scores(tokenized_query)
sorted_keyword_indices = np.argsort(keyword_scores)[::-1]

# Create a dictionary to map doc index to its keyword rank
keyword_ranks = {doc_id: rank + 1 for rank, doc_id in enumerate(sorted_keyword_indices)}
```

```python
# Create a set of all unique document IDs
all_doc_ids = set(semantic_indices) | set(keyword_ranks.keys())

# Define the constant k for the RRF formula
k = 60 

# Calculate RRF scores for all documents
reranked_scores = {}
for doc_id in all_doc_ids:
    # Get the ranks, defaulting to infinity if the document wasn't in a list
    semantic_rank = semantic_ranks.get(doc_id, float('inf'))
    keyword_rank = keyword_ranks.get(doc_id, float('inf'))

    # Calculate the RRF score
    score = (1 / (k + semantic_rank)) + (1 / (k + keyword_rank))
    reranked_scores[doc_id] = score

# Sort the documents by their new RRF score in descending order
final_ranking = sorted(reranked_scores.items(), key=lambda item: item[1], reverse=True)
```

```python
top_n = 5 # Display the top 5 results
print("\n--- Hybrid Search Results (Reciprocal Rank Fusion) ---")
for rank, (doc_index, score) in enumerate(final_ranking[:top_n]):
    if doc_index in loaded_doc_store:
        print(f"Rank {rank + 1} (Score: {score:.4f}):")
        print(f"  Document: {loaded_doc_store[doc_index]}")
        
        # Optional: show the individual ranks from each search
        semantic_rank = semantic_ranks.get(doc_index, 'N/A')
        keyword_rank = keyword_ranks.get(doc_index, 'N/A')
        print(f"  Individual Ranks -> Semantic: {semantic_rank}, Keyword: {keyword_rank}")
        print("-" * 20)
    else:
        # Handle cases where a doc_id might not be in the store (shouldn't happen with this setup)
        print(f"Rank {rank + 1} (Doc ID: {doc_index}) not found in document store.")
```
